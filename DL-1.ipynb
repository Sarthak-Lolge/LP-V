```python
# 1. Import
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler

# 2. Load dataset
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()

print("Shapes:", train_X.shape, train_Y.shape)

# 3. Check for NaN (IMPORTANT)
print("NaN in train_X:", np.isnan(train_X).sum())
print("NaN in train_Y:", np.isnan(train_Y).sum())

# 4. Normalize
scaler = StandardScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)

# 5. Build model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(train_X.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

# 6. Compile
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 7. Train
model.fit(train_X, train_Y, epochs=50, batch_size=16)

# 8. Evaluate
loss, mae = model.evaluate(test_X, test_Y)
print("MAE:", mae)
```
